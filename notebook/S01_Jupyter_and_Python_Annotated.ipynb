### Control flow

- if and the ternary operator
- Checking conditions - what evaluates as true/false?
- if-elif-else
- while
- break, continue
- pass

```python
if 1 + 1 == 2:
    print("Phew!")    
```

```python
'vegan' if 1 + 1  == 2 else 'carnivore'
```

```python
'vegan' if 1 + 1  == 3 else 'carnivore'
```

```python
if 1+1 == 3:
    print("oops")
else:
    print("Phew!")
```

```python
for grade in [94, 79, 81, 57]:
    if grade > 90:
        print('A')
    elif grade > 80:
        print('B')
    elif grade > 70:
        print('C')
    else:
        print('Are you in the right class?')
```

```python
i = 10
while i > 0:
    print(i)
    i -= 1    
```

```python
for i in range(1, 10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
for i in range(1, 10):
    if i % 2 == 0:
        break
    print(i)
```

```python
for i in range(1, 10):
    if i % 2 == 0:
        pass
    else:
        print(i)
```

### Loops and comprehensions

- for, range, enumerate
- lazy and eager evaluation
- list, set, dict comprehensions
- generator expression

```python
for i in range(1,5):
    print(i**2, end=',')
```

```python
for i, x in enumerate(range(1,5)):
    print(i, x**2)
```

```python
for i, x in enumerate(range(1,5), start=10):
    print(i, x**2)
```

```python
range(5)
```

```python
list(range(5))
```

#### Comprehensions

```python
[x**3 % 3 for x in range(10)]
```

```python
{x**3 % 3 for x in range(10)}
```

```python
{k: v for k, v in enumerate('abcde')}
```

```python
(x**3 for x in range(10))
```

```python
list(x**3 for x in range(10))
```
